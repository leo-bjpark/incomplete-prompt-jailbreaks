```python
from modules import load_base_model, generate_response
from modules import make_incomplete_chat_user_prompt, make_incomplete_chat_response_prompt
from datasets import load_dataset
```

```python
# model, tokenizer = load_base_model("google/gemma-3-4b-it")
model, tokenizer = load_base_model("meta-llama/Llama-3.1-8B-Instruct")
```

```python
incomplete_ds = load_dataset("leo-bjpark/incomplete-prompt-jailbreak", "completion_benchmark", split="train")
chat_template_ds = load_dataset("leo-bjpark/incomplete-prompt-jailbreak", "chat_template_benchmark", split="train")
```

```python
incomplete_prompts = incomplete_ds['prompt']
chat_user_incomplete_prompts = [make_incomplete_chat_user_prompt(tokenizer, prompt) for prompt in incomplete_prompts]
chat_response_incomplete_prompts = [make_incomplete_chat_response_prompt(tokenizer, v['user_request'], v['response_prefill']) for v in chat_template_ds]
```

```python
print("--------------------------------")
print("incomplete_prompts[0]", incomplete_prompts[0])
print("--------------------------------")
print("chat_user_incomplete_prompts[0]", chat_user_incomplete_prompts[0])
print("--------------------------------")
print("chat_response_incomplete_prompts[0]", chat_response_incomplete_prompts[0])
print("--------------------------------")
```

```python
index=0
responses = generate_response(model, tokenizer, [incomplete_prompts[index],
                                                 chat_user_incomplete_prompts[index],
                                                 chat_response_incomplete_prompts[index]],  max_new_tokens=128)
```
